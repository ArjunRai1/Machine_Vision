```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/OIP.jpg")
plt.imshow(img)
```

```python
print("Dimensions of image: ", img.shape)
print("Shape of image: ", )
```

```python
b, g, r = cv2.split(img) #Split the rgb values
print(b)
print(g)
print(r)
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) #Convert to gray image
plt.imshow(gray_image, 'gray') #Render the image as gray image
```

```python
print("Shape of image: ", gray_image.shape)
print("Dimensions: ", gray_image.ndim) #Print dimensions of image
```

```python
ret, thresh = cv2.threshold(gray_image, 120, 255, 0)

#Display original, grayscale and binary
plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title("Original Image"), plt.axis('off')
plt.subplot(132), plt.imshow(gray_image, cmap='gray'), plt.title("Grayscale Image"), plt.axis('off')
plt.subplot(133), plt.imshow(thresh, cmap='gray'), plt.title("Binary Image"), plt.axis('off')
```
